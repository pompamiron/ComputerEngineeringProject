# K-mean

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_scaled = pd.DataFrame(mms.fit_transform(X), 
                         columns=X.columns, 
                         index=X.index)
X_scaled
```

```python
from sklearn.cluster import KMeans
cls = KMeans(n_clusters=2, n_jobs=-1)
cls.fit(X_scaled)
```

```python
centroid = pd.DataFrame(cls.cluster_centers_, columns=X.columns)

import seaborn as sns
sns.heatmap(centroid)

print(centroid)
```
